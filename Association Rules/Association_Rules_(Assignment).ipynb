## **Data Preprocessing**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_excel('/content/Online retail.xlsx', header=None)
df
```

```python
df = df.rename(columns={0: 'Transactions'})
df
```

## **Association Rule Mining**

### **Apply Apriori Algorithm**

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df_encoded = df['Transactions'].str.get_dummies(sep=',')
df_encoded
```

```python
# Apply Apriori algorithm to find frequent itemsets
freq_items = apriori(df_encoded, min_support=0.02, use_colnames=True)

# Display frequent itemsets
print(freq_items)
```

### **Generate Association Rules**

```python
# Generate association rules
rules = association_rules(freq_items, metric="lift", min_threshold=1.0)

# Display the association rules sorted by lift
print(rules.sort_values(by='lift', ascending=False).head())
```

## **Analysis and Interpretation**

###**Interpretation**:

**1. Support**: Indicates how frequently the antecedent and consequent appear together in the dataset.

*For example:*

*The rule (spaghetti) -> (ground beef) has a support of 0.039195, meaning these items appear together in approximately 3.92% of all transactions.*

**2. Confidence**: Measures how often the rule is true.

## **Interview Questions**

### **1. What is lift and why is it important in Association rules?**

**Lift**: Measures how much more often items A and B occur together than expected if they were independent.  
**Importance**: Helps identify meaningful associations between items by filtering out rules where items are independent or occur together by chance.

### **2. What is support and confidence? How do you calculate them?**

**Support**: Measures the frequency of occurrence of an itemset in the dataset.   
**Calculation**: Number of transactions containing the itemset divided by total transactions.   
**Confidence**: Measures the reliability or strength of a rule.   
**Calculation**: Support of itemset A and B divided by support of itemset A.

### **3. What are some limitations or challenges of Association rules mining?**

**High Dimensionality**: Increases computational complexity and rule generation.  
**Sparse Data**: Can lead to infrequent itemsets and less meaningful associations.  
**Parameter Selection**: Choosing appropriate thresholds for support, confidence, and lift affects rule quality.  
**Interpretation**: Rules may require domain knowledge for actionable insights.   
**Overfitting**: Mining large datasets without validation may result in non-generalizable rules.
