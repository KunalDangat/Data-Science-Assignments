# **Data Visualization**

```python
# Plot histograms for each numerical column.
import seaborn as sns
import matplotlib.pyplot as plt

df1 = pd.DataFrame(numerical_df)

# Plot histograms for each column
for column in df1.columns:
    sns.histplot(df1[column], bins = 10)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

***Analyze the distribution (e.g., skewness, presence of outliers) and provide inferences.***

**1) Volume**: The **positive skewness** in the volume column indicates that the distribution is skewed to the **right**. This means that there are **more occurrences** of **lower volume** values compared to higher ones, with a **few values** of **very high volume**. In terms of **outliers**, you might expect to find values on the **higher end of the volume** that are significantly **larger than the rest**.

**2) Average Price**: **Similar to volume**, the **Right skewness** in the average price column suggests that the **majority** of **average prices** are **lower**, with a long tail of higher average prices. **Outliers** in this column could be **prices** that are **higher than** the **rest of the data**.

**3) Total Sales Value**: The **positive skewness** here implies that **most** of the **sales values are lower**, with a **smaller** number of **transactions** contributing to **higher sales** values. **Outliers** might be **exceptionally high sales** values **compared** to the **bulk of the data**.

**4) Discount Rate**: Since the distribution is **left-skewed**, the **majority** of **transactions** are having **higher discount** rates, possibly indicating frequent application of discounts or promotions. **Outliers** in this context would likely be **transactions** with unusually **low discount rates** compared to the rest of the data.

**5) Discount Amount**: **Positive skewness** in the discount amount column suggests that **smaller discounts** for **most transactions**, but there are **some transactions** with significantly **larger discounts**. **Outliers** could represent **transactions** where **unusually high discounts** were applied.

**6) Net Sales Value**: With **positive skewness** in net sales value, it's likely that **most transactions** result in **lower net sales** values, but there are a **few transactions** with **much higher values**. **Outliers** here would be **transactions** with **exceptionally high** net **sales** values compared to the majority.

```python
#	Create boxplots for numerical variables to identify outliers and the interquartile range.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df2 = pd.DataFrame(numerical_df)

fig, subplot = plt.subplots(3, 2, figsize=(15, 20))
subplot = subplot.flatten()

for i, column in enumerate(df2.columns):
    sns.boxplot(df2[column], ax=subplot[i])
    subplot[i].set_title(f'Boxplot of {column}',fontsize=16)
    subplot[i].set_xlabel(column,fontsize=14)
    subplot[i].set_ylabel('Value',fontsize=14)

plt.tight_layout()
plt.show()
```

Interquartile Range (IQR): The length of the box in each box plot represents the interquartile range (IQR), which covers the middle 50% of the data. A wider box indicates a larger spread of values within the dataset.

# **Standardization of Numerical Variables**

**Explain the concept of standardization (z-score normalization).**

Standardization, also known as z-score normalization, is a statistical technique used to transform a dataset so that it has a mean of 0 and a standard deviation of 1.

```python
#	Standardize the numerical columns using the formula: z=x-mu/sigma
import pandas as pd

means = numerical_df.mean()

std_dev = numerical_df.std()

standardized_df = (numerical_df - means) / std_dev

print(standardized_df)
```

```python
# Show before and after comparisons of the data distributions.
# Before Standardization
numerical_df
```

```python
# After Standardization
standardized_df
```

**After standardization:**

1.The mean of both columns becomes 0.   
2.The standard deviation of both columns becomes 1.   
3.Each value in the dataset is transformed to represent how many standard deviations it is from the mean.
